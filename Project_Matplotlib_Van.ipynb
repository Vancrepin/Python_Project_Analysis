## TP Matplotlib

Nous travaillerons avec l'ensemble de données de vente d'un détaillant en ligne. Les données sont collectées sur une période de trois ans : 2012 à 2015. Elles contiennent les informations sur les ventes réalisées par l'entreprise.

Les produits capturés appartiennent à trois catégories :

     Meubles
     Fournitures de bureau
     Technologie


```python
# importer numpy et  matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creer un array avec des categories de produits 
product_category = np.array(['Furniture', 'Tech', 'Office_Supply'])
sales = np.array([4110451.90, 4744557.50, 3787492.52])
```

## 1- Bar Graph: Tracer les ventes pour
chaque catégorie de produits

- Un graphique à barres utilise des barres pour afficher des comparaisons entre des catégories de données.

- Un axe aura généralement des valeurs numériques ou des mesures,
- L'autre décrira les types de catégories comparées ou les dimensions


```python
plt.bar(product_category, sales, width = 0.8 , )
plt.title("Sales across Product Catégories", fontsize=18, color="green")
plt.xlabel("product_categories", fontsize=18, color="green")
plt.ylabel(" sales ", fontsize=18, color="green")
plt.xticks(rotation = 45)
plt.show()
```


## 2- Scatter Chart - Tracer les ventes en fonction des profits 
Scatter plots (nuages de points) sont utilisés lorsque vous souhaitez montrer la relation entre deux faits ou mesures.

- Le profit est mesuré dans chaque catégorie de produits

```python
product_category = np.array(['Furniture', 'Tech', 'Office_Supply','Furniture', 'Tech', 'Office_Supply'])
sales = np.array([4110451.90, 4744557.50, 3787492.52,2980451.90, 3744557.50, 2787492.52])
profit = np.array([1213.46, 1814.13, 1485.7,2213.46, 2814.13, 2485.7])
```

```python

plt.scatter(profit, sales, s = 50)
plt.xlabel("profit", color = "red")
plt.ylabel("sales", color = "red")
plt.title( "sales Versus Profits of product Categories", fontsize = 18, color = "green" )
plt.show()
```

```python
plt.scatter(profit[product_category == "Furniture"], sales[product_category == "Furniture"], 
            c= 'Green', s = 150, label="Furniture")

plt.scatter(profit[product_category == "Tech"], sales[product_category == "Tech"], 
            c= 'Yellow', s = 100, label="Tech" )

plt.scatter(profit[product_category == "Office_Supply"], sales[product_category == "Office_Supply"], 
            c= 'Cyan',  s = 50, label="Office_Supply" )

# Adding and formatting title
plt.title("Sales versus Profits across Product Categories\n", fontsize= 20, fontweight= 5, color = 'Green')

# Labeling Axes
plt.xlabel("Profit",fontsize= 15, fontweight= 5, color = 'Brown')
plt.ylabel("Sales", fontsize= 15, fontweight= 5, color = 'Brown')

plt.legend()

plt.show()
```

## 3- Scatter Chart - Tracer les ventes en fonction des profits de maniere plus sophistiquee

```python
# Extracting the indices of each category.
furniture = np.where(product_category == "Furniture")
tech = np.where(product_category == "Tech")
office_supply = np.where(product_category == "Office_Supply")
 
# The categories en plot independantly.
plt.figure(figsize=(5,3))
plt.scatter(profit[furniture], sales[furniture], s=100, c="green", label="Furniture")
plt.scatter(profit[tech], sales[tech], s=50, c="yellow", label="Tech")
plt.scatter(profit[office_supply], sales[office_supply], s=30, c="skyblue", label="Office_Supply")
plt.xticks(np.arange(1200, 2900, 200))
#plt.yticks(np.arange(2.75e6, 5e6, 0.25e6))
plt.legend(loc="upper right")
plt.show()
```

# 4- Line Chart - Plot Ventes en 2015

A line chart ou un graphique linéaire ou un graphique linéaire ou un graphique à courbes est un type de graphique qui affiche des informations sous la forme d'une série de points de données appelés « marqueurs » reliés par des segments de ligne droite. Plus couramment utilisé avec les données temporelles.

```python
months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
sales = np.array([241268.56, 184837.36, 263100.77, 242771.86, 288401.05, 401814.06, 258705.68, 456619.94, 481157.24, 422766.63, 555279.03, 503143.69])
```

```python
from matplotlib.ticker import FuncFormatter
plt.figure(figsize=(6,5))
x = months
y = sales # Sine Graph
plt.plot(x,y,color='blue', linewidth=1.0,linestyle='-')
plt.scatter(x,y,color='blue', marker ='*')
plt.title("Ventes en 2015", fontsize = 15, color = "green" )
plt.xlabel("Mois", color = "r")
plt.ylabel("Ventes", color = "r")
plt.xticks(rotation = 45)
plt.yticks(range(0, 550001, 50000))
plt.ylim(bottom = 0)
plt.show()
```

# 5- Histogramme

Un histogramme est un tableau de fréquence qui enregistre l'occurrence d'une entrée ou d'un élément dans un ensemble de données. Il est utile lorsque vous souhaitez comprendre la distribution d'une série donnée.

```python
age=np.array([21,23,22,24,24,23,23,22,23,24])
#plotting a hitogram
plt.hist(age, bins=10, color='yellowgreen', edgecolor='b')
plt.title('Histogram',fontweight='bold')
```
